```python
grs=[]
```

```python
start=0
end=1

for page in range(start,end):
    try:
        gr=driver.find_elements(By.XPATH,'//span[@name="nv"][2]')
        for i in gr:
            grs.append(i.text)
    except NoSuchElementException:#handling no such element exception
            grs.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            grs.append('NA')
```

```python
start=0
end=5
for page in range(start,end):
    try:
        sl=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
        for i in sl[0:100]:
            sl_no.append(i.text)
    except NoSuchElementException:#handling no such element exception
        sl_no.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
        sl_no.append('NA')
    
    try:
        mn=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]')
        for i in mn:
            movie_name.append(i.text.split('. ')[1])
    except NoSuchElementException:#handling no such element exception
        movie_name.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
        movie_name.append('NA')
    try:
        du=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
        for i in du:
            dur.append(i.text.split(' ')[0])
    except NoSuchElementException:#handling no such element exception
            dur.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            dur.append('NA')
    
    try:
        ye=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
        for i in ye:
            year.append(i.text)
    except NoSuchElementException:#handling no such element exception
            year.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            year.append('NA')
    try:
        rat=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
        for i in rat:
            rating.append(i.text)
    except NoSuchElementException:#handling no such element exception
            rating.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            rating.append('NA')
    
    try:
        meta=driver.find_elements(By.XPATH,'//span[@class="metascore  favorable"]')
        for i in meta:
            metascore.append(i.text)
    except NoSuchElementException:#handling no such element exception
            metascore.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            metascore.append('NA')
    try:
        vot=driver.find_elements(By.XPATH,'//span[@name="nv"][1]')
        for i in vot:
            votes.append(i.text)
    except NoSuchElementException:#handling no such element exception
            votes.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            votes.append('NA')
    try:
        gr=driver.find_elements(By.XPATH,'//span[@name="nv"][2]')
        for i in gr:
            gross.append(i.text)
    except NoSuchElementException:#handling no such element exception
            gross.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            gross.append('NA')
    
      
    try:
        gen=driver.find_elements(By.XPATH,'//span[@class="genre"]')
        for i in gen:
            genre.append(i.text)
    except NoSuchElementException:#handling no such element exception
            genre.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            genre.append('NA')
    
    
    try:
        cer=driver.find_elements(By.XPATH,'//span[@class="certificate"]')
        for i in cer:
            certif.append(i.text)
    except NoSuchElementException:#handling no such element exception
            certif.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            certif.append('NA')
    try:
        dire=driver.find_elements(By.XPATH,'//p//a[1]')
        for i in dire:
            director.append(i.text)
    except NoSuchElementException:#handling no such element exception
            director.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            director.append('NA')
    try:
        sta=driver.find_elements(By.XPATH,'//p//a[2]')
        for i in sta:
            stars.append(i.text)
    except NoSuchElementException:#handling no such element exception
            stars.append('NA')
    except StaleElementReferenceException:#handling Stale element exception
            stars.append('NA')
    try:
        dir_star=driver.find_elements(By.XPATH,'//p[3]')
        for i in dir_star:
            d_s.append(i.text.split('|')[0].split(':')[1])
            starall.append(i.text.split('|')[1].split(':')[1])
    except NoSuchElementException:
        d_s.append('NA') 
        starall.append('NA')    

    
   
    
    NEXT_tag=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    NEXT_tag.click()
    time.sleep(5)
```
